```python
#
df = pd.read_csv('BTC_DATA.csv',low_memory=False)
display(df.sample(10))
```

```python
#Data Preprocessing:
```

```python
#check for missing values in the data using pandas isnull(), describe() function to get some initial statistics

print('A) Missing values - ')
display(df.isnull())

print('B) Total missing values per column - ')
display(df.isnull().sum())

print('C) Total missing values - ')
display(df.isnull().sum().sum())



print("Describing dataframe")
display(df.describe())

```

```python
#Provide variable descriptions. Types of variables etc.

print('Types of vars - ')
display(df.dtypes)
```

```python
#Check the dimensions of the data frame.
print('Dimension of df - ')
display(df.ndim)
```

```python
# Data Formatting and Data Normalization: 
```

```python
# Summarize the types of vars by checking the data types (i.e., character, numeric, integer, factor, and logical) 
# of the variables in the data

print('Character - ')
display(df.select_dtypes(include='object'))

print('Numeric - ')
display(df.select_dtypes(include='number'))

print('Integer - ')
display(df.select_dtypes(include='int64'))

print('Logical - ')
display(df.select_dtypes(include='bool'))

             
```
